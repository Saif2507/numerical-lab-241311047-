```python
import numpy as np
import matplotlib.pyplot as plt

# Example polynomial: f(x) = x^3 + 2x^2 + 3x + 1
def f(x):
    return x**3 + 2*x**2 + 3*x + 1

# Data points
x = np.array([0, 1, 2, 3, 4], dtype=float)
y = f(x)

n = len(x)
diff_table = np.zeros((n, n))
diff_table[:, 0] = y

# Calculate forward differences
for j in range(1, n):
    for i in range(n - j):
        diff_table[i][j] = diff_table[i+1][j-1] - diff_table[i][j-1]

# Print forward difference table 
print("\nPolynomial Function: f(x) = x^3 + 2x^2 + 3x + 1")
print("\nForward Difference Table (Up to 4th Order):\n")
print("x\tf(x)\tΔ¹\tΔ²\tΔ³\tΔ⁴")

for i in range(n):
    print(f"{x[i]:.1f}", end="\t")
    for j in range(n - i):
        print(f"{diff_table[i][j]:.1f}", end="\t")
    print()

# Optional: Plot the polynomial
plt.plot(x, y, marker='o', color='green', label='f(x) = x³ + 2x² + 3x + 1')
plt.title("Polynomial Function Plot")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
