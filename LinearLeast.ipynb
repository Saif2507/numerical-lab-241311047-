```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([1,2,3,4,5])
y = np.array([0.6, 2.4, 3.5, 4.8, 5.7])
 
m = len(y)
 
sum_x = np.sum(X)
sum_y = np.sum(y)
sum_xy =np.sum(X*y)
sum_x2 =np.sum(X**2)
 
x_mean = np.mean(X)
y_mean = np.mean(y)
 
 

a1 = ((m * sum_xy) - (sum_x * sum_y)) / ((m * sum_x2) - (sum_x**2))
a0 = y_mean - a1 * x_mean
 
print("Slope (a1):", a1)
print("Intercept (a0):", a0)
 

y_pred = a0 + a1 * X
 

plt.scatter(X, y, color='blue', label='Data points')  
plt.plot(X, y_pred, color='red', label=f'Best fit line: y = {a0:.2f} + {a1:.2f}x')
 
plt.title('Linear Least Squares Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
